**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
df = df.groupby(['Area Name','Year']).mean()
df.head()
```

```python
df = df.reset_index()
df.head()
```

```python
def ratio(past,future):
    boro_dict = {}
    for boro in df['Area Name']:
        if boro not in boro_dict:
            area = df[df['Area Name'] == boro]
            past_price = float(area['Avg Price'][area['Year'] == past])
            future_price = float(area['Avg Price'][area['Year'] == future])
            ratio = future_price/past_price   # instructions say divide '98 price by '18 price but harder to read, taking reciprocal
            boro_dict[boro] = ratio
    boro_df = pd.DataFrame(boro_dict.items(), columns=['Boroughs','Ratio']).sort_values('Ratio',ascending=False).reset_index(drop=True)
    boro_top = boro_df[:5]
    return boro_top
```

```python
ratio(1995,2015) # top 5 boroughs with the greatest increases in housing price from 1995 to 2015
```

```python
for x in range(1995,2000):
    print(ratio(x,x+20))
```

Observation: the average prices are more disparate the earlier the start year

```python
final_top = ratio(1998,2018)   # top 5 boroughs with the greatest increases in housing price specifically from 1998 to 2018
final_top
```

```python
ax = final_top.plot.bar(x='Boroughs', y='Ratio', rot=0)
plt.xticks(rotation=90)
plt.ylim(bottom=4)
plt.show()
```

```python
def counting():  # create a function that makes a flat list of all the top 10 ratios for applicable 20 year spans
    counting = []
    
    for x in range(1995,2000):
        create_df = ratio(x,x+20)
        top_10 = create_df['Boroughs'].to_list()
        lol_count = counting.extend(top_10)
    return counting
```

```python
from collections import Counter
count_list = Counter(counting())
count_list.most_common(5)
```

```python
plt.bar(count_list.keys(), count_list.values())
plt.xticks(rotation=90)
plt.show()
```
